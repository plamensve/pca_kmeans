# Profiler
---
This notebook analyzes the results of the models and aims to better understand the underlying structure of the data. The goal is to examine why the models assign a given user to a particular cluster and to interpret the characteristics that define each segment.


#### *PURCHASES_FREQUENCY*

The purchase frequency coefficient for **Cluster 3** is above **0.9**, and **Cluster 1** also shows a very high value, close to **0.9**.

In contrast, **Cluster 0** and **Cluster 2** exhibit significantly lower purchase frequency values, indicating that users in these groups make purchases much less often.

---

#### *PURCHASES_TRX*

The **number of purchase transactions** is another feature that clearly separates the clusters.

**Cluster 3** once again dominates in terms of transaction count, followed by **Cluster 1**.
The remaining two clusters (**Cluster 0** and **Cluster 2**) show the lowest number of transactions.

---

#### *PURCHASES_INSTALLMENTS_FREQUENCY*

A similar pattern appears in this feature.
**Cluster 3** and **Cluster 1** have the highest frequency of purchases made in installments, while **Cluster 0** and **Cluster 2** remain at considerably lower levels.

---

#### *Interpretation*

After examining several of the top features from the **feature importance ranking**, it becomes clear that **Cluster 3 stands out strongly from the others**. This suggests that the model can assign users to this cluster more easily.

The only misclassifications in the test data (exactly **three cases**) involve **Cluster 1**, which may be explained by the fact that the values of the examined features for these users were very close to those of other clusters.
